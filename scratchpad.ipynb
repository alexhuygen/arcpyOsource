```
import time

def coffee_timer(duration_seconds):
    print(f"Starting coffee timer for {duration_seconds} seconds...")
    for i in range(duration_seconds, 0, -1):
        print(f"Time remaining: {i} seconds", end='\r')
        time.sleep(1)
    print("\nCoffee is ready! Enjoy your coffee!        ")

# Set your desired brewing time in seconds here
# For example, 240 seconds for 4 minutes
brewing_time = 240
coffee_timer(brewing_time)
```
